```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
reps = pd.read_csv('house_of_rep_election.csv')
```

```python
reps.tail(5)
```

```python
reps['party'].value_counts()
```

```python
reps = reps.drop(columns=['state_fips', 'state_cen', 'state_ic', 'office', 'version', 'mode', 'candidate'])
```

```python
reps['party'].describe()
```

```python
reps['percent_vote'] = reps['candidatevotes']/reps['totalvotes']
```

```python
reps.tail(5)
```

```python
reps = reps[ reps['stage'] == 'gen']
reps = reps[ reps['party'].isin(['democrat', 'republican']) ]
reps = reps.drop(columns=['candidatevotes', 'totalvotes', 'stage'])
```

```python
reps.tail(5)
```

```python
reps_new = pd.pivot_table(reps, index = ['year', 'state', 'state_po', 'district'], columns = 'party', values = 'percent_vote')
reps_new = reps_new.reset_index()
```

```python
 reps_new.tail(5)
```

```python
reps_new['district'] = reps_new['district'].astype(str)
reps_new['rep_district'] = reps_new['state_po'] + "-" + reps_new['district']
reps_new = reps_new.fillna(0)
```

```python
reps_new.head(5)
```

```python
pacific = reps_new[ reps_new['state_po'].isin(['WA', 'OR', 'CA']) ]
```

```python
republican = pacific.pivot("rep_district", "year", "republican").fillna(0)
```

```python
fig, ax = plt.subplots(figsize=(30,30)) 

hm =sns.heatmap(republican,   ax=ax, cmap = sns.color_palette("vlag"))
#hm = sns.heatmap(democrat)
```

```python
test_district = reps_new[ reps_new['rep_district'] == 'MD-3']
plt.plot(test_district['year'], test_district['democrat'], label = "Democrat")
plt.plot(test_district['year'], test_district['republican'], label = "Republican")
plt.xlabel("Year")
plt.ylabel("% Vote")
plt.legend()
plt.show()
```
